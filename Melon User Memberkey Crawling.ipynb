```python
#라이브러리 불러오기
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
import pandas as pd
```

```python
#멜론 사이트 열기
dr = webdriver.Chrome('C:/chromedriver.exe')
dr.get('https://www.melon.com/chart/index.htm')
dr.find_element(By.CLASS_NAME,'btn_chart_f').click()
time.sleep(1)
```

```python
#조건 설정
dr.find_element(By.CLASS_NAME,'tab03').click()
time.sleep(1)
dr.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[2]/span/label').click()
time.sleep(1)
dr.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[10]/span/label').click()
time.sleep(1)
dr.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[1]/span/label').click()
time.sleep(1)
dr.find_element(By.CLASS_NAME,'btn_b26').click()
```

```python
memberkeys=[]
links=[]
song_list=dr.find_element(By.TAG_NAME,'tbody').find_elements(By.TAG_NAME,'tr')
for song in song_list[:10]:
    link=song.find_element(By.CLASS_NAME,'btn.btn_icon_detail').get_attribute('href')
    links.append(link)
```

```python
#멤버키 수집
for link in links:
    dr.execute_script(link)
    time.sleep(1)
    url=dr.current_url
    users=dr.find_elements(By.CLASS_NAME,'user_id.ellipsis.d_cmtpgn_user')
    for user in users:
        key=user.get_attribute('data-member-key')
        if key not in memberkeys:
            memberkeys.append(key)
    page=2
    while True:
        try:
            last=dr.find_element(By.CLASS_NAME,'btn_last.d_cmtpgn_navigator').get_attribute('href')
            dr.get(url+'#cmtpgn=&pageNo='+str(page)+'&sortType=0&srchType=2&srchWord=')
            time.sleep(1)
            users=dr.find_elements(By.CLASS_NAME,'user_id.ellipsis.d_cmtpgn_user')
            for user in users:
                key=user.get_attribute('data-member-key')
                if key not in memberkeys:
                    memberkeys.append(key)
            page+=1
        except:
            print('마지막 페이지입니다.')
            break
```

```python
for member in memberkeys:
    try:
        member=int(member)
    except:
        memberkeys.remove(member)
```
