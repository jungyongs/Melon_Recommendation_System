```python
# csr matrix 만들기 위해 indexing

# 고유한 유저, 아티스트를 찾아내는 코드
user_unique = data2['user_id'].unique()
title_unique = data2['title_artist'].unique()

# 유저, 아티스트 indexing 하는 코드
user_to_idx = {v:k for k,v in enumerate(user_unique)}
title_to_idx = {v:k for k,v in enumerate(title_unique)}
```

```python
data3=data2
data3['user_id']=data3['user_id'].map(user_to_idx.get)
data3['title_artist']=data3['title_artist'].map(title_to_idx.get)
data3
```

```python
from scipy.sparse import csr_matrix

csr_data = csr_matrix((data3.quantity, (data3.user_id, data3.title_artist)))
csr_data
```

```python
# Sparsity: 얼마나 비어있나?
matrix_size = csr_data.shape[0]* csr_data.shape[1]
num_like =len(csr_data.nonzero()[0])
sparsity = 100 * (1 - (num_like / matrix_size))
sparsity
```

```python
from implicit.als import AlternatingLeastSquares
import numpy as np
```

```python
# Implicit AlternatingLeastSquares 모델의 선언
# 1. factors : 유저와 아이템의 벡터를 몇 차원으로 할 것인지
# 2. regularization : 과적합을 방지하기 위해 정규화 값을 얼마나 사용할 것인지 
# 3. use_gpu : GPU를 사용할 것인지 
# 4. iterations : epochs와 같은 의미입니다.
als_model = AlternatingLeastSquares(factors=100, regularization=0.1, use_gpu=False, iterations=15, dtype=np.float32)
```

```python
# 모델 훈련
als_model.fit(csr_data)
```

```python
# 비슷한 노래 찾기
def get_similar_music(favorite_music):
    similar_music=[]
    title_id = title_to_idx[favorite_music]
    ids, scores = als_model.similar_items(title_id, N=20)
    for pair in zip(ids, scores):
        similar_music.append(pair)
    idx_to_title = {v:k for k,v in title_to_idx.items()}
    
    return [idx_to_title[i[0]] for i in similar_music]
```

```python
get_similar_music('사랑하기 때문에_유재하')
```

```python
# 노래 추천
def get_recommendations(your_name):
    user = user_to_idx[your_name]
    recommendations=[]
    ids, scores = als_model.recommend(user, csr_data[user], N=20)
    for pair in zip(ids, scores):
        recommendations.append(pair)
    idx_to_title = {v:k for k,v in title_to_idx.items()}
    
    return [idx_to_title[i[0]] for i in recommendations]
```
