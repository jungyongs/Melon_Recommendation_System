```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
import pandas as pd
```

```python
#멤버키 불러오기
df=pd.read_csv('memberkey2.csv')
user=df['memberkey'].values.tolist()
user=user[:499]
```

```python
#수집할 노래,가수,앨범 리스트 생성
memberkey=[]
title=[]
singer=[]
album=[]
```

```python
#데이터 크롤링
dr = webdriver.Chrome('C:/chromedriver.exe')
for key in user:
    dr.get('https://www.melon.com/mymusic/main/mymusicmainother_list.htm?memberKey='+str(key))
    dr.find_element(By.XPATH,'//*[@id="songType"]').click()
    time.sleep(1)
    song_list=dr.find_element(By.TAG_NAME,'tbody').find_elements(By.TAG_NAME,'tr')
    for song in song_list:
        memberkey.append(key)
        title.append(song.text.split('\n')[3])
        singer.append(song.text.split('\n')[4])
        album.append(song.text.split('\n')[5])
    try:
        last=dr.find_element(By.CLASS_NAME,'btn_last').get_attribute('href')
        result = re.sub(r'[^0-9]', '', last)
        for i in range(21,int(result)+1,20):
            dr.execute_script('javascript:pageObj.sendPage('+str(i)+');')
            time.sleep(1)
            song_list=dr.find_element(By.TAG_NAME,'tbody').find_elements(By.TAG_NAME,'tr')
        for song in song_list:
            memberkey.append(key)
            title.append(song.text.split('\n')[3])
            singer.append(song.text.split('\n')[4])
            album.append(song.text.split('\n')[5])
    except:
        print('마지막 페이지입니다.')
```

```python
#수집한 리스트 데이터프레임 변환
df_song=pd.DataFrame(zip(memberkey,title,singer,album),columns=['memberkey','title','singer','album'])
df_song
```

```python
#데이터프레임 파일로 변환
df_song.to_csv('likelist.csv',index=False)
```

```python
#좋아요 노래 목록1 데이터 불러오기
df1=pd.read_csv('./Datasets/likelist.csv')
df1
```

```python
#좋아요 노래 목록2 데이터 불러오기
df2=pd.read_csv('./Datasets/likelist2.csv')
df2.columns=['keynumber','title','artist','album']
df2
```

```python
#데이터 합치기
df3=pd.concat([df1,df2])
df3=df3.reset_index(drop=True)
df3=df3.dropna()
df3
```

```python
#노래 제목이 같은 노래 구분을 위해 노래_가수로 된 새로운 컬럼 생성
df3['title_artist']=df3['title']+'_'+df3['artist']
df3
```

```python
#목록에서 좋아요 누른 사람이 1명인 노래 전처리
df3['quantity']=[1]*201227
df3.groupby('title_artist')['quantity'].count()
```

```python
df4=pd.DataFrame(df3.groupby('title_artist')['quantity'].count())
df4=df4[df4['quantity']==1]
df4
```

```python
#목록에서 좋아요 누른 사람이 1명인 노래 삭제
delete=df3[df3['title_artist'].isin(df4.index)].index
df5=df3.drop(delete)
df5.reset_index(inplace=True,drop=True)
df5
```

```python
#데이터프레임 파일로 저장
df5.to_csv('likelist3.csv',index=False)
```
